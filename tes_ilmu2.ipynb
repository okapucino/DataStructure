```python
import numpy as np

arr = np.array( [70, 65, 85, 90, 66] )
```

```python
print("data array: ")
print(arr)
```

```python
nRendah = np.min(arr)
print("Nilai terendah: ")
print(nRendah)

nTinggi = np.max(arr)
print("Nilai tertinggi: ")
print(nTinggi)

nRata = np.mean(arr)
print("Nilai rata-rata: ")
print(nRata)


```
